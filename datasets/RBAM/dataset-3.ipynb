```python
import pandas as pd
import numpy as np
```

```python
politicalArg = pd.read_csv("../../rawData/political-argumentation/full_dataset.tsv", sep="\t")
politicalArg = politicalArg.rename(columns=
  {
    "argument1": "argSrc",
    "argument2": "argTrg"
  })

politicalArg = politicalArg.drop(columns=["pair_id", "source_arg_1", "source_arg_2"])

politicalArg['datasetSource'] = "political-argumentation"
politicalArg = politicalArg.replace("no_relation", "neither")
politicalArg.head(5)
```

```python
argGpt = pd.read_csv("../../rawData/ArGPT/Data/Data V2/argument_relations.csv", index_col=0)
argGpt = argGpt.rename(columns=
  {
    "source_tokens": "argSrc",
    "target_tokens": "argTrg",
    "labels" : "relation"
  })
argGpt = argGpt.drop(columns=["text_id", "source_id", "target_id", "split"])
argGpt['datasetSource'] = "argGpt"
argGpt['topic'] = "unknown"
argGpt = argGpt.replace(np.nan, "neither")
argGpt.head(5)
```

```python
argMicrotext = pd.read_csv("../../processedData/arg-microtexts-2-RBAM/data/mergedCorpusRelations.csv", index_col=0)
argMicrotext = argMicrotext.rename(columns=
  {
    "relationType": "relation",
  })
argMicrotext['datasetSource'] = "arg-microtexts"
argMicrotext['topic'] = "unknown"
argMicrotext.head(5)
```

```python
kialo = pd.read_csv("../../processedData/kialo-2-RBAM/kialoPairs.csv", index_col=0)
kialo.head(5)
```

```python
RBAMdataset = pd.concat([politicalArg, argGpt, argMicrotext, kialo])
```

```python
RBAMdataset
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(RBAMdataset, RBAMdataset["relation"])
```

```python
X_res
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, _, _ = train_test_split(X_res, X_res.relation, test_size=0.2, random_state=3854)
X_train, X_validation, _, _ = train_test_split(X_train, X_train.relation, test_size=0.0625, random_state=3854)
```

```python
X_train.to_csv("RBAM-dataset-3-TRAIN.csv")
X_test.to_csv("RBAM-dataset-3-TEST.csv")
X_validation.to_csv("RBAM-dataset-3-EVAL.csv")
```
